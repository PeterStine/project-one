```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
pth=Path('data','violence_2016_clean.xls')
df=pd.read_excel(pth)
df=df.rename(columns={'Violent\ncrime1':'violent crime per 100k inhabitants'})
df.head()
```

```python
final_table_columns = ['State', 'violent crime per 100k inhabitants']
df = df.drop(columns=[col for col in df if col not in final_table_columns])
```
