```python
import numpy as np
import pandas as pd
from pathlib import Path
import re
```

```python
def remove_chars(s):
    return re.sub('[^0-9]+', '', s)
def remove_digits(s):
    return re.sub('[^\sa-zA-Z]+', '', s)
```

```python
pthlead=Path('../data','lead_pipes_clean.csv',encoding='utf')
dflead=pd.read_csv(pthlead)
dflead.drop('LSLs per 100k people',axis=1,inplace=True)
dflead.drop('State Population',axis=1,inplace=True)
dflead.drop('Total LSLs (high estimate)',axis=1,inplace=True)
dflead['State']=dflead['State'].apply(remove_digits)
# dflead['State']=dflead['State'].str.encode('ascii', 'ignore').str.decode('ascii')
dflead['State']=dflead['State'].str.encode('ascii', 'replace').str.decode('ascii')
dflead['State']=dflead['State'].str.replace('?',' ')
# dflead['State']=dflead['State'].replace('?',' ',regex=True)
# dflead['State']=dflead['State'].str.replace(r"~[a-zA-Z]",'')
# dflead.set_index('State',inplace=True)
dflead.head()
dflead
# dflead.info()
```

```python
pthviolence=Path('../data','violence_2016_clean.csv')
dfviolence=pd.read_csv(pthviolence)
# dfviolence.set_index('State',inplace=True)
dfviolence['State']=dfviolence['State'].str.title()
dfviolence['State']=dfviolence['State'].apply(remove_digits)
dfviolence['State']=dfviolence['State'].str.encode('ascii', 'ignore').str.decode('ascii')
# dfviolence['State']=dfviolence['State'].str.replace(r"~[a-zA-Z]",'')
# dfviolence['State'].replace('5','',regex=True,inplace=True)
# dfviolence['State'].replace('6','',regex=True,inplace=True)
dfviolence.head()
dfviolence.info()
dfviolence
```

```python
# dflead.info()
dfmerged=dflead.merge(dfviolence,how='outer')
dfmerged=dfmerged.dropna(how='any')
dfmerged
```

```python
pthout=Path('../data','merged_data.csv')
dfmerged.to_csv(pthout,index=False)
```
