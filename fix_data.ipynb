```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
pth=Path('data','lead_pipes.csv')
df=pd.read_csv(pth)
df.head()
```

```python
df['State']=df['State'].str.lstrip('State')
df['Total LSLs (high estimate)']=df['Total LSLs (high estimate)'].str.lstrip('Total LSLs (high estimate)')
df['State Population']=df['State Population'].str.lstrip('State Population')
# df['LSLs per 100k people']=df['LSLs per 100k people'].str.lstrip('LSLs per 100k people')
df.head()
```

```python
df['Total LSLs (high estimate)']=df['Total LSLs (high estimate)'].replace(',','',regex=True)
df['State Population']=df['State Population'].replace(',','',regex=True)
df['LSLs per 100k people']=df['LSLs per 100k people'].replace('LSLs per 100k people ','',regex=True)
df['LSLs per 100k people']=df['LSLs per 100k people'].replace(',','',regex=True)
df.head()
```

```python
df['Total LSLs (high estimate)']=pd.to_numeric(df['Total LSLs (high estimate)'])
df['State Population']=pd.to_numeric(df['State Population'])
df['LSLs per 100k people']=pd.to_numeric(df['LSLs per 100k people'])
df.head()
df.info()
```

```python
df.to_csv('data/lead_pipes_clean.csv',index=False)
```
